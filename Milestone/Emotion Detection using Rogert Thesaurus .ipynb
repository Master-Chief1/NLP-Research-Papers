```python
import nltk
from nltk.corpus import wordnet as wn
from collections import defaultdict
```

```python
# Initialize the emotion categories from Roget's Thesaurus
roget_categories = defaultdict(list)
for synset in wn.all_synsets():
    for lemma in synset.lemmas():
        if lemma.antonyms():
            continue
        roget_categories[synset.pos()].append((lemma.name(), synset.name().split('.')[0]))
```

```python
# Map words to their associated emotions
text = "The weather is calm"
word_emotions = defaultdict(list)
for word in nltk.word_tokenize(text):
    word_synsets = wn.synsets(word)
    for synset in word_synsets:
        pos = synset.pos()
        if pos in roget_categories:
            for lemma in synset.lemmas():
                if lemma.name() in [w[0] for w in word_emotions[word]]:
                    continue
                for category in roget_categories[pos]:
                    if lemma.name() in category[0]:
                        word_emotions[word].append(category)
```

```python
# Print the emotions associated with each word
for word in word_emotions:
    print(word, word_emotions[word])
    print("\n")
```
